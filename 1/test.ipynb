```python
import matplotlib.pyplot as plt

# Test accuracy for each model
models = ['ResNet18', 'VGG', 'CNN', 'ANN']
test_acc = [85.89, 82.98, 62.85, 38.32]
colors = ['blue', 'red', 'green', 'orange']

# Sort models based on test accuracy
sorted_indices = sorted(range(len(test_acc)), key=lambda k: test_acc[k], reverse=True)
sorted_models = [models[i] for i in sorted_indices]
sorted_acc = [test_acc[i] for i in sorted_indices]

# Bar chart with ResNet18 highlighted
plt.figure(figsize=(12, 8))
bars = plt.bar(sorted_models, sorted_acc, color=colors)
plt.title('Model Test Accuracy Comparison')
plt.xlabel('Model')
plt.ylabel('Test Accuracy (%)')
plt.ylim(30, 100)  # Set the y-axis limit to start from 30%

# Annotate accuracy values and draw lines
for i, bar in enumerate(bars):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 1, f'{bar.get_height():.2f}%', fontsize=9)
    plt.axhline(y=bar.get_height(), color='gray', linestyle='--', xmax=0.92)

plt.show()
```
