```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('pizza.CSV')
x = data['size']
y = data['price']
plt.scatter(x, y)
plt.grid()
plt.axis([0, 25, 0, 25])
plt.title('Pizza price plotted against diameter')
plt.xlabel('Diameter in inches')
plt.ylabel('Price in dollars')
plt.show()
```
